```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup 
import sqlite3 
```

```python
url = 'https://en.everybodywiki.com/100_Most_Highly-Ranked_Films'
```

```python
db_name = 'Movies.db'
```

```python
table_name = 'Top_25_Movies'
```

```python
csv_name = 'top_25_films.csv'
```

```python
df = pd.DataFrame(columns=['Film', 'Year', 'Rotten Tomatoes Top 100'])
```

```python
count = 0
```

```python
html_page = requests.get(url).text
html_page
```

```python
data = BeautifulSoup(html_page, 'html.parser')
data
```

```python
tables = data.find_all('tbody')
tables 
```

```python
rows = tables[0].find_all('tr')
rows
```

```python
for row in rows: 

    if count<25:

        col = row.find_all('td')

        if len(col)!=0:

            data_dict = {'Film': col[1].contents[0],
                         'Year': col[2].contents[0],
                         'Rotten Tomatoes Top 100': col[3].contents[0]}
            
            df1 = pd.DataFrame(data_dict, index=[0])
            df = pd.concat([df, df1], ignore_index=True)

            count += 1

    else:
        break 
```

```python
df
```

```python
df['Year'].dtype
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.loc[df['Year'] >= 2000] 
```

```python
df.to_csv(csv_name)
```

```python
conn = sqlite3.connect(db_name)
```

```python
df.to_sql(table_name, conn, if_exists='replace', index=False)
```

```python
conn.close()
```
